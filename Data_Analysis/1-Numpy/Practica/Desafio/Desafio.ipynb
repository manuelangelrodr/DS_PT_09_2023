como se cual debo hacer muy simple introduce tu fecha de cumpleaños en la siguiente funcion

```python
def cual_es_mi_reto(fecha: str) -> str:
    """
    Args:
    ------
    fecha: str el formato de la fecha es DD-MM-YYYY
    Returns:
    -------
    str: el numero del reto
    """
    try:
        return int(fecha.split('-')[0]) % 3
    except:
        return 'La fecha tiene que ser en este formato DD-MM-YYYY'
```

```python
import random

import numpy as np

tablero = np.full((10, 10), " ")

print(tablero)

def barco(tablero, longitud):
    '''
    Mediante esta función colocamos un barco de 4 posiciones en el tablero.

    Input:
    tablero: le indicamos el tamaño del tablero con el que estamos jugando

    Output:

    Barco de 4 posiciones en una dirección aleatoria
    '''
    while True:
        orientacion = random.choice(['N', 'S', 'E', 'O']) #se asigna una dirección aletoria de 4 posibles al barco de 4 posiciones

        if orientacion == 'N':
            fila = random.randint(longitud -1, 9)
            columna = random.randint(0, 9)
            posicion = [(fila - i, columna) for i in range(longitud)]

        elif orientacion == 'S':
            fila = random.randint(0, 9 - longitud + 1)
            columna = random.randint(0, 9)
            posicion = [(fila + i, columna) for i in range(longitud)]

        elif orientacion == 'E':
            fila = random.randint(0, 9)
            columna = random.randint(0, 9 - longitud + 1)
            posicion = [(fila, columna + i) for i in range(longitud)]

        else:
            fila = random.randint(0, 9)
            columna = random.randint(longitud - 1, 9)
            posicion = [(fila, columna - i) for i in range(longitud)]

        # Verificar si las posiciones son válidas y no hay otros barcos en ellas
        if all(0 <= fila < 10 and 0 <= columna < 10 and tablero[fila, columna] == ' ' for fila, columna in posicion):
            for fila, columna in posicion:
                tablero[fila, columna] = 'O'
            break


barco1 = int(input("Elija el tamaño del barco 1 (1 - 4): "))
barco(tablero, barco1)

barco2 = int(input("Elija el tamaño del barco 2 (1 - 4): "))
barco(tablero, barco2)

barco3 = int(input("Elija el tamaño del barco 3 (1 - 4): "))
barco(tablero, barco3)

barco4 = int(input("Elija el tamaño del barco 4 (1 - 4): "))
barco(tablero, barco4)

barco5 = int(input("Elija el tamaño del barco 5 (1 - 4): "))
barco(tablero, barco5)

print(tablero)


def disparar(tablero):
    '''
    Comienza el juego de los barquitos con el tablero de 10 x 10
    '''

    while True:
        print(tablero)
        
    
        # Ingresar las coordenadas del disparo 0 - 9
        fila = int(input("Ingrese fila (0-9): "))
        columna = int(input("Ingrese columna (0-9): "))
    
        
        if tablero[fila, columna] == "O":
            print("Tocado")
            tablero[fila, columna] = "X"  

        elif tablero[fila, columna] == "X":
            print("Ya habías elegido esta casilla")
            tablero[fila, columna] = "X"

        else:
            print("Fallo")
            tablero[fila, columna] = "-"  
    
        # Si todos los barcos se han hundido
        if np.sum(tablero == "O") == 0:
            print("¡Has hundido todos los barcos!")
            break


disparar(tablero)
```
